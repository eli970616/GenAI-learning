```python
# Import necessary libraries
import matplotlib.pyplot as plt  # Import Matplotlib's pyplot module as plt
import numpy as np  # Import NumPy library for numerical operations

# Generate sample data
x = np.linspace(0, 10, 100)  # Generate 100 evenly spaced points from 0 to 10
y = np.sin(x)  # Compute the sine of each x value
categories = ['A', 'B', 'C']  # Define category labels
values = [5, 7, 3]  # Define values for each category
data = np.random.randn(1000)  # Generate 1000 random numbers from a standard normal distribution for histogram

# 1. Plotting a Line Chart
plt.figure(figsize=(10, 6))  # Create a new figure with a size of 10x6 inches
plt.plot(x, y, color='blue', linestyle='-', marker='o', label='Sine Wave')  # Plot the sine wave with specified color, line style, marker, and label
plt.title('Line Chart Example')  # Set the chart title
plt.xlabel('X-axis')  # Set the X-axis label
plt.ylabel('Y-axis')  # Set the Y-axis label
plt.legend()  # Display the legend
plt.grid(True)  # Add grid lines
plt.show()  # Display the plot

# 2. Plotting a Scatter Plot
plt.figure(figsize=(10, 6))  # Create a new figure
plt.scatter(x, y, color='red', marker='x', label='Scatter Points')  # Plot scatter points with specified color, marker, and label
plt.title('Scatter Plot Example')  # Set the chart title
plt.xlabel('X-axis')  # Set the X-axis label
plt.ylabel('Y-axis')  # Set the Y-axis label
plt.legend()  # Display the legend
plt.grid(True)  # Add grid lines
plt.show()  # Display the plot

# 3. Plotting a Bar Chart
plt.figure(figsize=(8, 5))  # Create a new figure
plt.bar(categories, values, color=['green', 'orange', 'purple'])  # Plot a bar chart with specified colors for each bar
plt.title('Bar Chart Example')  # Set the chart title
plt.xlabel('Category')  # Set the X-axis label
plt.ylabel('Value')  # Set the Y-axis label
plt.show()  # Display the plot

# 4. Plotting a Histogram
plt.figure(figsize=(10, 6))  # Create a new figure
plt.hist(data, bins=30, color='skyblue', edgecolor='black')  # Plot a histogram with 30 bins, specified color, and black edges
plt.title('Histogram Example')  # Set the chart title
plt.xlabel('Value')  # Set the X-axis label
plt.ylabel('Frequency')  # Set the Y-axis label
plt.show()  # Display the plot

# 5. Plotting a Pie Chart
sizes = [50, 30, 20]  # Define the sizes of each pie slice
labels = ['Part A', 'Part B', 'Part C']  # Define labels for each pie slice
colors = ['gold', 'lightcoral', 'lightskyblue']  # Define colors for each pie slice
explode = (0.1, 0, 0)  # Define the fraction of the radius with which to offset each wedge (explode the first slice)
plt.figure(figsize=(8, 8))  # Create a new figure
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)  # Plot the pie chart with specified parameters
plt.title('Pie Chart Example')  # Set the chart title
plt.axis('equal')  # Ensure the pie chart is drawn as a circle
plt.show()  # Display the plot

# 6. Creating Subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # Create a 2x2 grid of subplots and return the figure and axes objects

# Subplot 1: Line Chart
axs[0, 0].plot(x, y, 'r-', label='Sine Wave')  # Plot the sine wave in the first subplot with red solid line
axs[0, 0].set_title('Subplot 1: Line Chart')  # Set the title for the first subplot
axs[0, 0].legend()  # Display the legend in the first subplot
axs[0, 0].grid(True)  # Add grid lines to the first subplot

# Subplot 2: Scatter Plot
axs[0, 1].scatter(x, y, color='green', label='Scatter Points')  # Plot scatter points in the second subplot with green color
axs[0, 1].set_title('Subplot 2: Scatter Plot')  # Set the title for the second subplot
axs[0, 1].legend()  # Display the legend in the second subplot
axs[0, 1].grid(True)  # Add grid lines to the second subplot

# Subplot 3: Bar Chart
axs[1, 0].bar(categories, values, color='purple')  # Plot a bar chart in the third subplot with purple color
axs[1, 0].set_title('Subplot 3: Bar Chart')  # Set the title for the third subplot
axs[1, 0].set_xlabel('Category')  # Set the X-axis label for the third subplot
axs[1, 0].set_ylabel('Value')  # Set the Y-axis label for the third subplot

# Subplot 4: Histogram
axs[1, 1].hist(data, bins=20, color='orange', edgecolor='black')  # Plot a histogram in the fourth subplot with 20 bins
axs[1, 1].set_title('Subplot 4: Histogram')  # Set the title for the fourth subplot
axs[1, 1].set_xlabel('Value')  # Set the X-axis label for the fourth subplot
axs[1, 1].set_ylabel('Frequency')  # Set the Y-axis label for the fourth subplot

plt.tight_layout()  # Automatically adjust subplot parameters to give specified padding
plt.show()  # Display all subplots

# 7. Using Axes Objects for More Control
fig, ax = plt.subplots(figsize=(10, 6))  # Create a new figure and axes object
ax.plot(x, y, label='Sine Wave', color='blue', linewidth=2)  # Plot the sine wave on the axes with specified label, color, and line width
ax.set_title('Line Chart Using Axes Object')  # Set the chart title
ax.set_xlabel('X-axis')  # Set the X-axis label
ax.set_ylabel('Y-axis')  # Set the Y-axis label
ax.legend()  # Display the legend on the axes
ax.grid(True)  # Add grid lines to the axes
plt.show()  # Display the plot

# 8. Adding Annotations and Text
plt.figure(figsize=(10, 6))  # Create a new figure
plt.plot(x, y, label='Sine Wave', color='blue')  # Plot the sine wave with specified label and color
plt.title('Line Chart with Annotations')  # Set the chart title
plt.xlabel('X-axis')  # Set the X-axis label
plt.ylabel('Y-axis')  # Set the Y-axis label
plt.legend()  # Display the legend
plt.grid(True)  # Add grid lines
plt.text(2, 0.5, 'Peak', fontsize=12, color='red')  # Add text annotation at position (2, 0.5) with specified font size and color
plt.annotate('Trough', xy=(3 * np.pi, -1), xytext=(7, -0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))  # Add an annotation with an arrow pointing to (3, -1)
plt.show()  # Display the plot

# 9. Saving the Plot as a File
plt.figure(figsize=(10, 6))  # Create a new figure
plt.plot(x, y, label='Sine Wave', color='blue')  # Plot the sine wave with specified label and color
plt.title('Save Plot Example')  # Set the chart title
plt.xlabel('X-axis')  # Set the X-axis label
plt.ylabel('Y-axis')  # Set the Y-axis label
plt.legend()  # Display the legend
plt.grid(True)  # Add grid lines
plt.savefig('sine_wave.png')  # Save the current figure as a PNG file named 'sine_wave.png'
plt.show()  # Display the plot

# 10. Using Different Colors and Styles
plt.figure(figsize=(10, 6))  # Create a new figure
plt.plot(x, y, color='#FF5733', linestyle='--', marker='s', label='Styled Line')  # Plot the sine wave with a specific color, dashed line, square markers, and label
plt.title('Colors and Styles Example')  # Set the chart title
plt.xlabel('X-axis')  # Set the X-axis label
plt.ylabel('Y-axis')  # Set the Y-axis label
plt.legend()  # Display the legend
plt.grid(True)  # Add grid lines
plt.show()  # Display the plot
```
